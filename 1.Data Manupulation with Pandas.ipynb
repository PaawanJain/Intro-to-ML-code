#### Find how much time it take to deliver a order

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
```

```python
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data['delivery_days']=data['Ship Date']-data['Order Date']
```

```python
data['delivery_days']
```

```python
data['delivery_days'].dtype
```

#### How many time has it taken more than 5 days from placing an order to shipping ?

```python
(data['delivery_days'] > '5 days').sum()
```

```python
data[(data['delivery_days'] > '5 days')].shape[0]
```

```python
a=((data['delivery_days'] > '5 days').sum()/(data.shape[0]))*100
```
